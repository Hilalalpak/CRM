```python
# Dt_Customer tipini object yerine datetime olarak değiştirelim.
df["Dt_Customer"] = df["Dt_Customer"].apply(lambda x: pd.to_datetime(x, dayfirst=True))


#Z verilerini çıkartalım.
dlt = ["Z_CostContact", "Z_Revenue"]
df.drop(dlt, axis=1, inplace=True)
```

```python
df['Education'].value_counts()
df["Marital_Status"].value_counts()
```

> Veriyi daha sade hale getirmek adına şu şekilde ilerleyeceğim;

- Year_Birth değerlerini kullanılır hale getirmek için müşterinin yaşını hesaplayalım.
- Bireylerin Marital_Status bilgilerinden anlamsız bulduğum absurd ve yolo değerlerini çıkartacağım.
- Aile içinde kaçar kişinin yaşadığını belirtecek şekilde Family_Size değişkeni oluşturalım.
- Çocukları ve gençleri toplam belirtmek için Children isminde değişken oluşturalım.
- Education için değerleri sadeleştirmek adına gruplayalım.
- Müşterinin iki yıl boyunca harcadığı toplam tutarı gösteren "Monetary" kolonu oluşturalım.
- Ve bazı gereksiz, anlamsız gördüğüm değerleri veriden çıkartacağım.

```python
# Müşterilerin bugünkü yaşını hesaplayalım.
df["Age"] = dt.datetime.today().year-df["Year_Birth"]


# Çocuk sayılarını birleştirelim.
df["Children"] = df["Kidhome"] + df["Teenhome"]


# Anlamsız değerleri çıkartalım. Aile içindeki toplam kişi sayısını bulalım.
df = df[(df["Marital_Status"] != "Absurd") & (df["Marital_Status"] != "YOLO")]

df["Marital_Status"].replace({"Single":1,"Together":2, "Married":2,"Divorced":1,"Widow":1,"Alone":1}, inplace=True)
df["Family_Size"] = df["Marital_Status"] + df["Children"]





# Eğitim seviyelerini gruplayalım.
df["Education"] = df["Education"].replace({"Basic":"Undergraduate",
                                           "2n Cycle":"Undergraduate", 
                                           "Graduation":"Graduate",
                                           "Master":"Postgraduate", 
                                           "PhD":"Postgraduate"})

# Kolon isimlerini sadeleştirelim. 
df = df.rename(columns={"MntWines": "Wines",
                        "MntFruits":"Fruits",
                        "MntMeatProducts":"Meat",
                        "MntFishProducts":"Fish",
                        "MntSweetProducts":"Sweet",
                        "MntGoldProds":"Gold"})


# Toplam ödemeleri hesaplayalım.
df["Monatery"] = df["Wines"] + df["Fruits"] + df["Meat"] + df["Fish"] + df["Sweet"] + df["Gold"]


delete = ["Year_Birth", "Marital_Status", "Kidhome", "Teenhome"]
df = df.drop(delete, axis=1)

print(" DESCRIBE ".center(100,'-'))
df.describe()
```

```python
# Aykırı değerleri traşlayalım.

def outlier_thresholds(dataframe, variable):
    quartile1 = dataframe[variable].quantile(0.15)
    quartile3 = dataframe[variable].quantile(0.85)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit.round(), up_limit.round()


def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
    
    
replace_with_thresholds(df,"Income")    
replace_with_thresholds(df,"Monatery")
replace_with_thresholds(df,"Age")
```

```python
# Recency ve Monetary değeri veride mevcut. F değerini hesaplayalım.
# Ardından 5 bölüme ayırarak puanlayalım. rfm isimli dataframe e gönderelim.

df["Frequency"] = df["NumWebPurchases"] + df["NumCatalogPurchases"] + df["NumStorePurchases"]



df["Recency_Score"] = pd.qcut(df["Recency"], 5, labels=[5, 4, 3, 2, 1])

df["Monatery_Score"] = pd.qcut(df["Monatery"], 5, labels=[1, 2, 3, 4, 5])

df["Frequency_Score"] = pd.qcut(df["Frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])

df["RF_SCORE"] = (df["Recency_Score"].astype(str) +
                    df["Frequency_Score"].astype(str))   # Monetary değeri çok fazla olası kombinasyon getireceği için değerlendirmeye katmadım.


df.head()
```

```python
# RFM analizini daha rahat değerlendirmek adına rfm isimlik dataframe e istediğimiz verileri alalım.
rfm = df[["ID", "Monatery", "Recency", "Frequency", "Monatery_Score", "Recency_Score", "Frequency_Score", "RF_SCORE"]]
rfm.head()
```

```python
# RF skorlarına göre segmentlere ayıralım.

seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'}

rfm['Segment'] = rfm['RF_SCORE'].replace(seg_map, regex = True)

rfm.head(10).style.background_gradient(cmap='Blues')
```

```python
# Diğer işlemler için df e segment kolonunu ekleyelim.
df["Segment"] = rfm["Segment"]
df.head()
```

```python
# Aynı Segmentteki değerlerin medyanına göre Income içindeki boş değerleri dolduralım.
df['Income'] = df.groupby('Segment')['Income'].transform(lambda x: x.fillna(x.median()))

# Kontrol edelim.
df.isnull().sum().sum()
```

> Segmentasyon sonucuna göre, müşterilerin önemli bir kısmı, At_Risk, Hibernating, About_to_Sleep gibi düşük katılım kategorilerinde gözüküyor.
Şirket, müşterilerin neden aktif olmadığını belirlemesi ve hedef kitleyi geri kazanmak için stratejiler geliştirmesi gerekiyor.

```python
# Grafik oluşturalım.

segment_counts = df['Segment'].value_counts()

plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, y="Segment", order=segment_counts.index)

total = len(rfm['Segment'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_width() / total)
    x = p.get_width() + 0.02
    y = p.get_y() + p.get_height() / 2
    ax.annotate(percentage, (x, y), ha='left', va='center', fontsize=10, color='black')


plt.xlabel('Count')
plt.ylabel('Segment')
plt.title('Segment Counts with Percentage')
plt.show()
```

```python
# RFM isimli df oluşturup, içerisine gerekli kolonları df'ten çekelim.
RFM = df[["ID", "Segment", "Monatery", "Frequency", "Recency", "Monatery_Score", "Frequency_Score", "Recency_Score", "RF_SCORE"]]

RFM.head()
```
